# Task 1: Exploring and Visualizing a Simple Dataset

## IRIS DATASET

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("IRIS.csv")
```

```python
# Inspect the dataset
print("Dataset Shape:", df.shape)       
print("\nColumns:", df.columns.tolist()) 
print("\nFirst 5 rows:")
print(df.head())                
```

```python
#Summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
#Visualization Section

# Set a style for plots
sns.set(style="whitegrid")
```

```python
# --- Scatter Plot ---
# Relationship between sepal length and sepal width
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species", palette="viridis")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
# --- Histogram ---
# Distribution of petal length
plt.figure(figsize=(6,4))
sns.histplot(data=df, x="petal_length", kde=True, bins=20, color='skyblue')
plt.title("Distribution of Petal Length")
plt.show()
```

```python
# --- Box Plot ---
# Spread of sepal length by species
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="species", y="sepal_length", palette="Set2")
plt.title("Box Plot of Sepal Length by Species")
plt.show()
```
